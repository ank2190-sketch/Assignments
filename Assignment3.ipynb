## Functions

### 1) Write a function to convert temperature from kelvin to celsius

```python
def kelvin_to_celsius(temp_k):
    """Convert temperature from Kelvin to Celsius"""
    return temp_k - 273.15
```

```python
def celsius_to_kelvin(temp_c):
    """Convert temperature from Celsius to Kelvin"""
    return temp_c + 273.15
```

```python
# Test kelvin to celsius
print(f"300 K = {kelvin_to_celsius(300)} °C")
print(f"273.15 K = {kelvin_to_celsius(273.15)} °C")

# Test celsius to kelvin
print(f"0 °C = {celsius_to_kelvin(0)} K")
print(f"25 °C = {celsius_to_kelvin(25)} K")
```

### 2) Write a function to convert temperature to fahrenheit

```python
def celsius_to_fahrenheit(temp_c):
    """Convert temperature from Celsius to Fahrenheit"""
    return (temp_c * 9/5) + 32
```

```python
def kelvin_to_fahrenheit(temp_k):
    """Convert Kelvin to Fahrenheit"""
    temp_c = kelvin_to_celsius(temp_k)
    return celsius_to_fahrenheit(temp_c)
```

### 4) Now write a function that converts from farenheit and uses a keyword argument to specify whether you want the output in celcius or kelvin

```python
def from_fahrenheit(temp_f, output_unit='celsius'):
    """ Convert temperature from Fahrenheit to Celsius or Kelvin """
    temp_c = (temp_f - 32) * 5/9
    
    if output_unit == 'celsius':
        return temp_c
    else: 
        return temp_c + 273.15
```

```python
print(from_fahrenheit(32))
print(from_fahrenheit(212))
```

### 5) Write a function that takes two arguments (feet and inches) and returns height in meters. Verify it gives sensible answers

```python
def feet_inches_to_meters(feet, inches):
    """Convert height from feet and inches to meters"""
    total_inches = (feet * 12) + inches
    meters = total_inches * 0.0254
    return meters
```

```python
print(f"6 feet 0 inches = {feet_inches_to_meters(6,0):.2f} meters")
```

```python
print(f"5 feet 4 inches = {feet_inches_to_meters(5,4):.2f} meters")
```

```python
print(f"5 feet 2 inches = {feet_inches_to_meters(5,2):.2f} meters")
```

### 6) Write a function takes one argument (height in meters) and returns two arguments (feet and inches)

```python
def meters_to_feet_inches (meters):
    """Convert Height from meters to Feet and Inches"""
    total_inches = meters / 0.0254
    feet = int(total_inches // 12)
    inches = total_inches % 12
    return feet, inches
```

```python
feet, inches = meters_to_feet_inches(1.83)
print(f"1.83 meters = {feet} feet {inches:.1f} inches")
```

```python
feet, inches = meters_to_feet_inches(1.53)
print(f"1.53 meters = {feet} feet {inches:.1f} inches")
```
